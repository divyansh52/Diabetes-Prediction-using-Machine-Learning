Some features like Glucose, BloodPressure, Skin Thickness, Insulin and BMI have their min values as 0. It means some of the data points in the datasets are missing so we need to impute the values later.

```python
# Visualising the datasets graphically

%matplotlib inline
import matplotlib.pyplot as plt
diabetes.hist(bins = 50, figsize = (20, 15))
```

```python
# Replacing the 0 values as Nan so that they can be imputed later
diabetes_copy = diabetes.copy(deep = True)
diabetes_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
# In order to fill in the missing values, using the median startegy of imputer to fill in the missing values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "median")
imputer.fit(diabetes_copy)
```

```python
X = imputer.transform(diabetes_copy) # Filling median in all columns which contains null values
```

```python
diabetes_tr = pd.DataFrame(X, columns = diabetes.columns) # Converting numpy array into dataframe
```

```python
diabetes_tr.describe().T
```

```python
# After  imputing the values with their missing values.

diabetes_tr.hist(bins = 50, figsize = (20, 15))
```

# Splitting the data in training and testing set

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(diabetes_tr, test_size = 0.2, random_state = 42)
print(f"Rows in train set : {len(train_set)} \nRows in test set : {len(test_set)}\n")
```

```python
# Copy the training set and prepare a corelation chart
diabetes = train_set.copy()
```

```python
corr_matrix = diabetes.corr()
```

```python
corr_matrix['Outcome'].sort_values(ascending = False)
```

```python
# Plotting corelation plots
from pandas.plotting import scatter_matrix
attributes = [ 'Glucose', 'BMI', 'Age', 'Outcome']
scatter_matrix(diabetes[attributes], figsize = (12, 8))
```

```python
#Separate the data and labels

X_train = train_set.drop("Outcome", axis=1)
X_test = test_set.drop("Outcome", axis=1) 

Y_train= train_set["Outcome"].copy()
Y_test= test_set["Outcome"].copy()
```

## Trained various models and compared their testing accuracy

```python
from sklearn import metrics
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import  LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
from sklearn.metrics import precision_score, recall_score
import seaborn as sns
```

```python
# Multinomial Naive Bayes Classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train,Y_train)
```

```python
# Predict the testing set 
pred_nb = nb_classifier.predict(X_test)

# Calculate the accuracy score
score = metrics.accuracy_score(Y_test,pred_nb)
print("Testing accuracy for Multinomial Naive Bayes is :"+str(score))

test_accuracy = []
models=[]

test_accuracy.append(score)
models.append("NB")
```

```python
cm_nb = metrics.confusion_matrix(Y_test, pred_nb)
precision_metric = precision_score(Y_test, pred_nb, average = "macro")
recall = recall_score(Y_test, pred_nb, average = "macro")
print(f"Confusion matrix for Multinomial Naive Bayes is :\n{cm_nb}")
print(f"Precision Score for Multinomial Naive Bayes is : {precision_metric}")
print(f"Recall Score for Multinomial Naive Bayes is : {recall}")
```

```python
plt.figure(figsize=(4,4))

group_names = ["True Neg","False Pos","False Neg","True Pos"]
               
group_counts = ["{0:0.0f}".format(value) for value in cm_nb.flatten()]

labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
LABELS = np.asarray(labels).reshape(2,2)



sns.heatmap(cm_nb, annot=LABELS, fmt='', cmap='Blues')

plt.title("Confusion matrix for Multinomial Naive Bayes")
plt.ylabel('True class')
plt.xlabel('Predicted class')
b, t = plt.ylim()
b += 0.5 
t -= 0.5 
plt.ylim(b, t) 
plt.show()
plt.show()
```

```python
# Logistic Regression Classifier
logr_classifier = LogisticRegression()
logr_classifier.fit(X_train,Y_train)
```

```python
# Predict the testing set
predicted_logr= logr_classifier.predict(X_test)

# Calculate the accuracy score: score
score = metrics.accuracy_score(Y_test, predicted_logr)
print("Testing accuracy of Logistic Regression is :"+str(score))

test_accuracy.append(score)
models.append("LR")
```

```python
cm_logr = metrics.confusion_matrix(Y_test, predicted_logr)
precision_metric_logr = precision_score(Y_test, predicted_logr, average = "macro")
recall_logr = recall_score(Y_test, predicted_logr, average = "macro")
print(f"Confusion matrix for Logistic Regression is :\n{cm_logr}")
print(f"Precision Score for Logistic Regression is : {precision_metric_logr}")
print(f"Recall Score for Logistic Regression is : {recall_logr}")
```

```python
plt.figure(figsize=(4,4))

group_names = ["True Neg","False Pos","False Neg","True Pos"]
               
group_counts = ["{0:0.0f}".format(value) for value in cm_logr.flatten()]

labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
LABELS = np.asarray(labels).reshape(2,2)



sns.heatmap(cm_logr, annot=LABELS, fmt='', cmap='Blues')

plt.title("Confusion matrix for Logistic Regression")
plt.ylabel('True class')
plt.xlabel('Predicted class')
b, t = plt.ylim()
b += 0.5 
t -= 0.5 
plt.ylim(b, t) 
plt.show()
plt.show()
```

```python
#SVM Classifier
rbf_svc = svm.SVC(kernel='rbf')
rbf_svc.fit(X_train,Y_train)
```

```python
# Predict the testing set 
pred_rbf = rbf_svc.predict(X_test)

# Calculate the accuracy score
score = metrics.accuracy_score(Y_test,pred_rbf)
print("Testing accuracy for SVM  is :"+str(score))

test_accuracy.append(score)
models.append("SVM")
```

```python
cm_svm = metrics.confusion_matrix(Y_test, pred_rbf)
precision_metric_svm = precision_score(Y_test, pred_rbf, average = "macro")
recall_svm = recall_score(Y_test, pred_rbf, average = "macro")
print(f"Confusion matrix for SVM is :\n{cm_svm}")
print(f"Precision Score for SVM is : {precision_metric_svm}")
print(f"Recall Score for SVM is : {recall_svm}")
```

```python
plt.figure(figsize=(4,4))

group_names = ["True Neg","False Pos","False Neg","True Pos"]
               
group_counts = ["{0:0.0f}".format(value) for value in cm_svm.flatten()]

labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
LABELS = np.asarray(labels).reshape(2,2)



sns.heatmap(cm_svm, annot=LABELS, fmt='', cmap='Blues')

plt.title("Confusion matrix for SVM")
plt.ylabel('True class')
plt.xlabel('Predicted class')
b, t = plt.ylim()
b += 0.5 
t -= 0.5 
plt.ylim(b, t) 
plt.show()
plt.show()
```

```python
# Random Forest Classifier
rf_clf = RandomForestClassifier(max_depth=3, random_state=0)
rf_clf.fit(X_train,Y_train)
```

```python
# Predict the testing set 
pred_rf = rf_clf.predict(X_test)

# Calculate the accuracy score
score = metrics.accuracy_score(Y_test,pred_rf)
print("Testing accuracy for Random Forest  is :"+str(score))

test_accuracy.append(score)
models.append("RF")
```

```python
cm_rf = metrics.confusion_matrix(Y_test, pred_rf)
precision_metric_rf = precision_score(Y_test, pred_rf, average = "macro")
recall_rf = recall_score(Y_test, pred_rf, average = "macro")
print(f"Confusion matrix for Random Forest is :\n{cm_rf}")
print(f"Precision Score for Random Forest is : {precision_metric_rf}")
print(f"Recall Score for Random Forest is : {recall_rf}")
```

```python
plt.figure(figsize=(4,4))

group_names = ["True Neg","False Pos","False Neg","True Pos"]
               
group_counts = ["{0:0.0f}".format(value) for value in cm_rf.flatten()]

labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
LABELS = np.asarray(labels).reshape(2,2)



sns.heatmap(cm_rf, annot=LABELS, fmt='', cmap='Blues')

plt.title("Confusion matrix for Random Forest")
plt.ylabel('True class')
plt.xlabel('Predicted class')
b, t = plt.ylim()
b += 0.5 
t -= 0.5 
plt.ylim(b, t) 
plt.show()
plt.show()
```

```python
plt.figure(figsize=(4,4))

group_names = ["True Neg","False Pos","False Neg","True Pos"]
               
group_counts = ["{0:0.0f}".format(value) for value in cm_dt.flatten()]

labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names,group_counts)]
LABELS = np.asarray(labels).reshape(2,2)



sns.heatmap(cm_dt, annot=LABELS, fmt='', cmap='Blues')

plt.title("Confusion matrix for Random Forest")
plt.ylabel('True class')
plt.xlabel('Predicted class')
b, t = plt.ylim()
b += 0.5 
t -= 0.5 
plt.ylim(b, t) 
plt.show()
plt.show()
```
